```python



import os
import random

# Define the folder containing the images
data_folder = '/content/drive/My Drive/APS360_Project/Dataset_RemovedWatermark'

# Define the classes and their initial image counts
class_counts = {'Resistor': 449,
                'LED': 564,
                'Transistor': 510,
                'Potentiometer': 463}

# Define the target count of images per class
target_count = 320

# Iterate through each class and delete excess images
for class_name, count in class_counts.items():
    # Calculate the number of images to delete
    delete_count = count - target_count

    # Check if there are excess images to delete
    if delete_count > 0:
        # Get the list of image files in the class folder
        class_folder = os.path.join(data_folder, class_name)
        image_files = os.listdir(class_folder)

        # Randomly select images to delete
        images_to_delete = random.sample(image_files, delete_count)

        # Delete the selected images
        for image in images_to_delete:
            image_path = os.path.join(class_folder, image)
            os.remove(image_path)

        # Update the count for the class
        class_counts[class_name] -= delete_count

print("Image deletion completed.")

```
